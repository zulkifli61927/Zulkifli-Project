```python
import requests
import json
import pandas as pd
```

```python
list_jam = [6]
for jam in list_jam:
  url = f"https://www.flightstats.com/v2/api/flights/dep/CGK/2023/3/24/{jam}?numHours=6&rqid=7dykrvd74tq"
  payload={}
  headers = {}
  response = requests.request("GET", url, headers=headers, data=payload)
  json_data = json.loads(response.text)
json_data['flights']
#print(response.text)
df = pd.json_normalize(json_data['flights'])
df
```

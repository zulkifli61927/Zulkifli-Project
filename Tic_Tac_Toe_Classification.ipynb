```python
#Melihat head pada data
TicTacToe_df = pd.read_csv("TicTacToe.csv")
TicTacToe_df.head()
```

```python
#definisikan X
X = TicTacToe_df[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9']].values
X[0:5]
```

```python
from sklearn import preprocessing

#ubah string b,x, dan o pada data ke integer
le_V1 = preprocessing.LabelEncoder()
le_V1.fit(['b', 'x', 'o'])
X[:, 0] = le_V1.transform(X[:, 0])

le_V2 = preprocessing.LabelEncoder()
le_V2.fit(['b', 'x', 'o'])
X[:, 1] = le_V2.transform(X[:, 1]) 

le_V3 = preprocessing.LabelEncoder()
le_V3.fit(['b', 'x', 'o'])
X[:, 2] = le_V3.transform(X[:, 2]) 

le_V4 = preprocessing.LabelEncoder()
le_V4.fit(['b', 'x', 'o'])
X[:, 3] = le_V4.transform(X[:, 3]) 

le_V5 = preprocessing.LabelEncoder()
le_V5.fit(['b', 'x', 'o'])
X[:, 4] = le_V5.transform(X[:, 4]) 

le_V6 = preprocessing.LabelEncoder()
le_V6.fit(['b', 'x', 'o'])
X[:, 5] = le_V6.transform(X[:, 5]) 

le_V7 = preprocessing.LabelEncoder()
le_V7.fit(['b', 'x', 'o'])
X[:, 6] = le_V7.transform(X[:, 6]) 

le_V8 = preprocessing.LabelEncoder()
le_V8.fit(['b', 'x', 'o'])
X[:, 7] = le_V8.transform(X[:, 7]) 

le_V9 = preprocessing.LabelEncoder()
le_V9.fit(['b', 'x', 'o'])
X[:, 8] = le_V9.transform(X[:, 8]) 

X[0:5]
```

```python
#definisikan y
y = TicTacToe_df['V10']
y[0:5]
```

```python
#mengecek data apa saja pada V10
TicTacToe_df["V10"].unique()
```

```python
#Definisikan X_train, X_test, y_train, dan y_test

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
#Cek data train
print(X_train.shape)
print(y_train.shape)
```

```python
#Cek data test
print(X_test.shape)
print(y_test.shape)
```

```python
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
TTTtree = DecisionTreeClassifier(criterion = 'entropy', max_depth = 4)
```

```python
TTTtree.fit(X_train, y_train)
```

```python
#data prediksi
predTree = TTTtree.predict(X_test)
predTree
```

```python
comparison = {"y_test" : y_test,
 "Predicted": predTree}
comp = pd.DataFrame(comparison)
comp
```

```python
#Hitung Akurasi prediksi
from sklearn.metrics import accuracy_score
print("Accuracy : ", accuracy_score(y_test, predTree) )
```

```python
searchResult = api.search_tweets(q='Indonesia world cup -RT', lang='en', count=500, tweet_mode='extended')
searchResult
```

```python
import datetime
import re
from textblob import TextBlob


fullTweet = []
for tweet in searchResult:
    tweetProperties = {}
    tweetProperties['date'] = tweet.created_at.strftime('%Y-%m-%d %H:%M:%S %Z')
    tweetProperties['name'] = tweet.user.screen_name
    tweetProperties['tweet'] = tweet.full_text
    tweetProperties['tweet'] = ' '.join(re.sub(r"(?::\d{2}[A-Z]?:|\r\n)+", " ", tweetProperties['tweet']).split())
    tweetProperties['tweet'] = re.sub(r'http\S+', '', tweetProperties['tweet'])
    analysisSentimentScore = TextBlob(tweetProperties['tweet']).sentiment.polarity
    if analysisSentimentScore < 0 :
        tweetProperties['sentiment'] = 'Negative'
    elif analysisSentimentScore > 0 :
        tweetProperties['sentiment'] = 'Positive'
    else :
        tweetProperties['sentiment'] = 'Netral'
    fullTweet.append(tweetProperties)

print(fullTweet)
```

```python
import pandas as pd
df =  pd.DataFrame(fullTweet)
df
```

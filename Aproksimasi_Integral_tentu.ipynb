```python
#import module yang dibutuhkan
from math import e, cos
from random import randint
import matplotlib.pyplot as plt
import numpy as np

#random bilangan bulat a dan b dari 1 hingga 9
a = randint(1,9)
b = randint(1,9)

print('untuk a = {} dan b = {}'.format(a,b))

#fungsi pada soal
def f(x):
  return (e**(-a*(x**2)))*abs(cos(b*x))+0.5

#membuat plot grafik fungsi f(x) dan integral f(x)
x = np.linspace(0,1,100)
y1 = e**(-a*(x**2))
y2 = np.cos(b*x)
y3 = e**(-a*((x**2)/4))
y4 = np.cos(b*x/2)
plt.plot(x, (y1*abs(y2))+0.5, '-r', label='f(x)')
plt.plot(x, (x/6)*(4+(4*y3*abs(y4))+(y1*abs(y2))), '-b', label='Integral dari f(x)') #(x/6)*(4+(4*y3*abs(y4))+(y1*abs(y2))) didapat dari integral f(x) dari 0 sampai x menggunakan metode simpson
plt.title('y = (e**(-a*(x**2)))*abs(cos(b*x))+0.5')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
plt.show()

#Nilai integral f(x) dari 0 sampai 1.
S = (1/6)*(f(0)+4*f(0.5)+f(1))
print('Luas Daerah = {} hektar'.format(S))
print('=================================================================================')
print('Luas tanah yang akan dibagi ke juki adalah {} hektar'.format(S))
```

```python
import datetime

dataID = []
for i in range (0,len(jsonDataID08['forecast'])):
    dateTime = datetime.datetime.fromtimestamp(jsonDataID08['forecast'][i]['timestamp'])
    plumeIndex = jsonDataID08['forecast'][i]['plume_index']
    plumeIndexLevel = jsonDataID08['forecast'][i]['plume_index_level']
    piNO2 = jsonDataID08['forecast'][i]['values']['NO2']['pi']
    piO3 = jsonDataID08['forecast'][i]['values']['O3']['pi']
    piPM10 = jsonDataID08['forecast'][i]['values']['PM10']['pi']
    piPM25 = jsonDataID08['forecast'][i]['values']['PM25']['pi']
    temperature = jsonDataID08['forecast'][i]['weather']['temperature']
    dataID.append([dateTime,plumeIndex,plumeIndexLevel,piNO2,piO3,piPM10,piPM25,temperature])

df = pd.DataFrame(dataID, columns=['Time','Plume Index','Plume Index Level','PI NO2','PI O3','PI PM10','PI PM25','Temperature'])
df
```

```python
df['Hour']=pd.to_datetime(df['Time']).apply(lambda x: x.hour)
```

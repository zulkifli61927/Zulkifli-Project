```python
#Import module yang dibutuhkan
import requests
import json

#url API pada website
url = "https://api.persibstore.id/store"

#payload pada url API di atas
payload = {
    
}
#headers pada url API di atas
headers = {
  "content-type": "application/json; charset=utf-8"}

#Request data
response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)
```

```python
#implementasikan json pada response.text di atas

json_data = json.loads(response.text)
```

```python
#Menkonstruksi Data frame
#import pandas untuk data frame
import pandas as pd

data=[]
rows = json_data['data']
for i in range(0, len(rows)):
    productid = json_data['data'][i]['images']['product_id']
    name = json_data['data'][i]['name']
    price = json_data['data'][i]['price']
    weight = json_data['data'][i]['weight']
    discount = json_data['data'][i]['discount']
    data.append(
        (productid, name, weight, price, discount)
    )

df = pd.DataFrame(data, columns=['Product ID', 'Nama Kaos', 'Berat (gram)', 'Harga awal (Rp)', 'Discount'])
df = df.replace('','0%')
df
```

```python
import numpy as np
from numpy.linalg import qr
```

```python
a = np.array([[1.5, 1, 0], 
              [1, 2, 1],
              [0, 1, 3]])
for i in range(20):
    q, r = qr(a)
    a = np.dot(r, q)
    print(f'Iteration {i+1}:')
    print('A', a)
```

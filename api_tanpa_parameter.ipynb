```python
import requests
```

```python
url = "https://api.covid19api.com/summary"

payload={}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)
print(response.text)
```

```python
# Ganti tipe ke JSON
import json

json_data = json.loads(response.text)
json_data['Countries']
```

```python
# Tabulasi
import pandas as pd

df = pd.json_normalize(json_data['Countries'])
df
```
